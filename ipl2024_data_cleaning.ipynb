### 2. Remove Duplicate Matches

```python
df.drop_duplicates(inplace= True)
```

### 3. Handle Missing Values

```python
df['player_of_the_match'].isnull().sum()
```

```python
df['player_of_the_match'] = df['player_of_the_match'].fillna(0)   #Filling NaN values with 0
```

```python
df['most_runs'].isnull().sum()
```

```python
a = df['second_score'].mean()
print(round(a, 0))

df['most_runs'] = df['most_runs'].fillna(a)
```

```python
df['most_wkts'].isnull().sum()
```

```python
df['most_wkts'] = df['most_wkts'].fillna(0)
```

### 4. Clean Team Names and Text Columns

```python
df['team1'] = df['team1'].str.strip()
```

```python
df['team1']
```

### 5. Fix Typos and Convert Data Types

```python
df['first_score'] = pd.to_numeric(df['first_score'])
```

```python
df['first_score']
```

### 6. Add new column (Win Margin)

```python
# Function to calculate win margin
def calculate_margin(row):
    if row['first_score'] > row['second_score']:
        return row['first_score'] - row['second_score']
    else:
        return 10 - row['second_wkts']

# Function to calculate win type
def calculate_type(row):
    if row['first_score'] > row['second_score']:
        return 'Runs'
    else:
        return 'Wickets'

# Apply functions to DataFrame
df['win_margin'] = df.apply(calculate_margin, axis=1)
df['win_type'] = df.apply(calculate_type, axis=1)
```
